```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
fig=plt.figure(figsize=(18,11))
xLabel = ['5%','10%','15%']
yLabel = ['0.85','0.90','0.95']

res = json.load(open('./Data/quantile_mitdata.json', 'r'))
length = {'gt_acceptors_achiral_binarypu': 838,'phosphatase_chiral_binarypu': 5424,'esterase_binarypu': 3076,'duf_binarypu': 274,'halogenase_NaBr_binarypu': 323}

data1 = []
for i in ('0.85','0.9','0.95'): 
    temp = []
    for j in (0.05,0.1,0.15):
        k = np.mean(res['phosphatase_chiral_binarypu'][i][str(int(length['phosphatase_chiral_binarypu']*j))])
        temp.append(k)  
    data1.append(temp)
ax1 = fig.add_subplot(232)

sns.heatmap(data1, cbar=True,square=True, cmap=sns.diverging_palette(230, 0))
ax1.set_yticks([0.5, 1.5, 2.5])
ax1.set_yticklabels(yLabel)
ax1.set_xticks([0.5, 1.5, 2.5])
ax1.set_xticklabels(xLabel)

plt.title("Phosphatase", fontsize=14)


data2 = []
for i in ('0.85','0.9','0.95'): 
    temp = []
    for j in (0.05,0.1,0.15):
        k = np.mean(res['esterase_binarypu'][i][str(int(length['esterase_binarypu']*j))])
        temp.append(k)
    data2.append(temp)
ax2 = fig.add_subplot(233)

sns.heatmap(data2, cbar=True,square=True, cmap=sns.diverging_palette(230, 0))
ax2.set_yticks([0.5, 1.5, 2.5])
ax2.set_yticklabels(yLabel)
ax2.set_xticks([0.5, 1.5, 2.5])
ax2.set_xticklabels(xLabel)
plt.title("Esterase", fontsize=16)



# GT  Glycosyltransferase
data3 = []
for i in ('0.85','0.9','0.95'): 
    temp = []
    for j in (0.05,0.1,0.15):
        k = np.mean(res['gt_acceptors_achiral_binarypu'][i][str(int(length['gt_acceptors_achiral_binarypu']*j))])
        temp.append(k)
    data3.append(temp)

ax = fig.add_subplot(234)
im = sns.heatmap(data3, cbar=True,square=True, cmap=sns.diverging_palette(230, 0))
ax.set_yticks([0.5, 1.5, 2.5])
ax.set_yticklabels(yLabel)
ax.set_xticks([0.5, 1.5, 2.5])
ax.set_xticklabels(xLabel)
plt.title("Glycosyltransferase", fontsize=16)


# BKACE
data4 = []
for i in ('0.85','0.9','0.95'): 
    temp = []
    for j in (0.05,0.1,0.15):
        k = np.mean(res['duf_binarypu'][i][str(int(length['duf_binarypu']*j))])
        temp.append(k)
    data4.append(temp)

ax3 = fig.add_subplot(235)
im = sns.heatmap(data4, cbar=True,square=True, cmap=sns.diverging_palette(230, 0))
ax3.set_yticks([0.5, 1.5, 2.5])
ax3.set_yticklabels(yLabel)
ax3.set_xticks([0.5, 1.5, 2.5])
ax3.set_xticklabels(xLabel)
plt.title("BKACE", fontsize=16)



# Halogenase
data5 = []
for i in ('0.85','0.9','0.95'): 
    temp = []
    for j in (0.05,0.1,0.15):
        k = np.mean(res['halogenase_NaBr_binarypu'][i][str(int(length['halogenase_NaBr_binarypu']*j))])
        temp.append(k)
    data5.append(temp)
    

ax4 = fig.add_subplot(236)

sns.heatmap(data5, cbar=True,square=True, cmap=sns.diverging_palette(230, 0))
ax4.set_yticks([0.5, 1.5, 2.5])
ax4.set_yticklabels(yLabel)
ax4.set_xticks([0.5, 1.5, 2.5])
ax4.set_xticklabels(xLabel)
plt.title("Halogenase", fontsize=16)


data6 = []

for d1, d2, d3, d4, d5 in zip(data1, data2, data3, data4, data5):
    lst_tmp = list()
    for tmp1, tmp2, tmp3, tmp4, tmp5 in zip(d1, d2, d3, d4, d5):
        lst_tmp.append(np.mean([tmp1, tmp2, tmp3, tmp4, tmp5]))
    data6.append(lst_tmp)

ax5 = fig.add_subplot(231)

sns.heatmap(data6, cbar=True,square=True, cmap=sns.diverging_palette(230, 0))
ax5.set_yticks([0.5, 1.5, 2.5])
ax5.set_yticklabels(yLabel)
ax5.set_xticks([0.5, 1.5, 2.5])
ax5.set_xticklabels(xLabel)
plt.title("All", fontsize=14)

ax5.set_ylabel("Thresholds", fontsize=16)
ax5.set_xlabel("Percentage of hidden TP", fontsize=16)


ax.tick_params(axis='both', which='major', labelsize=14)
ax1.tick_params(axis='both', which='major', labelsize=14)
ax2.tick_params(axis='both', which='major', labelsize=14)
ax3.tick_params(axis='both', which='major', labelsize=14)
ax4.tick_params(axis='both', which='major', labelsize=14)
ax5.tick_params(axis='both', which='major', labelsize=14)

plt.tight_layout()
plt.savefig('./S3.png', dpi=400)
```
