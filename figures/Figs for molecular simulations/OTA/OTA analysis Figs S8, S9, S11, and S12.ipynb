```python
import matplotlib.pylab as plt
import numpy as np
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

# Fig S8

```python
plt.figure(figsize=(9,3),dpi=400)

# from ./Data/mmgbsa.dat
mmpbsa_mean = [-37.1745,-38.0334,-36.6358,-47.8913,-40.0392,-34.3077,-12.7908,-23.331,-36.1273,-26.9866]
mmpbsa_std = [1.0733,0.517,0.6184,1.0235,0.4134,0.4625,0.947,0.5699,0.6807,0.7953]
width = 0.7

labels = ['S1P1','S1P2','S1P3','S1P4','S1P5','S1P6','S5P1','S5P2','S3P1','S5P3']
colors=['gray','red','gray','red','red','gray','gray','gray','gray','gray']
err_attr={"elinewidth":2,"capsize":3}

barlist = plt.bar(labels,mmpbsa_mean,width,yerr=mmpbsa_std,error_kw=err_attr)
for i, j in enumerate(barlist):
    j.set_color(colors[i])
    
plt.axhline(-38, linestyle='--')
plt.xticks(rotation=15)
plt.ylabel('$\Delta G$(Kcal/mol)',fontsize=15)
plt.title('Different Binding Poses MMGBSA',fontsize=15)
```

# Fig S9

```python
file = "./Data/S1P4/S1P4_per_residue_contribution.dat"
data = np.loadtxt(file)
plt.figure(figsize=(12,3),dpi=400)
colors1 = ['b','grey']

ind = np.arange(len(data[:,0]))
width=0.5
barlist = plt.bar(ind,data[:,1],width,yerr=data[:,2],error_kw=dict(ecolor='k', lw=0.5, capsize=1,capthick=1),align='center')

for i,j in enumerate(barlist):
    j.set_color('red')

ax = plt.gca()
plt.xlabel("Residue number",fontsize=15)
plt.ylabel("Total contri. (kcal/mol)",fontsize=15)
plt.ylim([-3,0])
plt.xlim([0,396])
plt.xticks(fontsize=10)
plt.title('S1P4 Per residue contribution',fontsize=15)
```

# Fig S11

```python
plt.figure(figsize=(12,3),dpi=400)
APO=np.loadtxt("./Data/OTA_APO/Apo.rmsf")
S1P2=np.loadtxt("./Data/S1P2/S1P2.rmsf")
S1P4=np.loadtxt("./Data/S1P4/S1P4.rmsf")
S1P5=np.loadtxt("./Data/S1P5/S1P5.rmsf")

lw=0.6
plt.plot(APO[:,0],APO[:,1],'k-',lw=lw)
plt.plot(S1P2[:,0]-1,S1P2[:,1],'b-',lw=lw)
plt.plot(S1P4[:,0]-1,S1P4[:,1],'g-',lw=lw)
plt.plot(S1P5[:,0]-1,S1P5[:,1],'r-',lw=lw)

plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.legend(['APO','S1P2','S1P4','S1P5'],fontsize=12)
plt.xlabel('Residue number',fontsize=12)
plt.ylabel('RMSF ($\AA$)',fontsize=12)
```

# Fig S12

```python
grids1 = np.linspace(0,1,7)
cs1 = ['red','orangered','w','w','w','limegreen','green']
cmap1 = LinearSegmentedColormap.from_list('mycmap',list(zip(grids1,cs1)))

APOcro=np.loadtxt('./Data/OTA_APO/Apocross-correlation.dat',comments="#")
S1P4cro0=np.loadtxt('./Data/S1P4/S1P4cross-correlation.dat',comments="#")
S1P4cro1=np.delete(S1P4cro0,1,axis=0)
S1P4cro=np.delete(S1P4cro1,1,axis=1)

plt.figure(figsize=(6,6),dpi=400)
plt.title('APO cross-correlation',fontsize=12)
plt.imshow(APOcro,cmap=cmap1,vmin=-1,vmax=1,origin='lower')
plt.colorbar()
```

```python
plt.figure(figsize=(6,6),dpi=400)
plt.title('S1P4 cross-correlation',fontsize=12)
plt.imshow(S1P4cro,cmap=cmap1,vmin=-1,vmax=1,origin='lower')
plt.colorbar()
```
